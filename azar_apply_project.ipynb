```python
def scraping(series):
    result = {
        'PhD': [],
        'Postdoctoral': [],
        'Master': [],
        'Germany': [],
        'Canada': [],
        'Australia': [],
        'USA': [],
        'Switzerland': [],
    }
    for item in series:
        for key in result:
            if key in item:
                result[key].append(item)
    return result
                
```

```python
scrp = scraping(positions['positions'])
```

```python
scrap = pd.DataFrame(scrp['PhD'], columns=['PhD'])
```

```python
for item in scrp:
    scrap[item] = pd.Series(scrp[item])
```

```python
scrap.to_csv('./clean_data.csv', sep=',', index=False, encoding='utf-8')
```
