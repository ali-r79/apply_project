```python
grades = ['Msc', 'Master', 'PhD', 'Postdoctoral', 'Postdoc', 'Research', 'Researcher', 'Job']
def grade_checker(text):
    words = text.split()
    for word in words:
        if word in grades:
            if word in ['Msc', 'Master']:
                return 'Msc'
            if word in ['Postdoctoral', 'Postdoc']:
                return 'Postdoc'
            if word in ['Research', 'Researcher']:
                return 'RA'
            return word
```

```python
positions['grade'] = positions['text'].transform(grade_checker)
```

```python
def scraping(series):
    result = {
        'PhD': [],
        'Postdoctoral': [],
        'Master': [],
        'Germany': [],
        'Canada': [],
        'Australia': [],
        'USA': [],
        'Switzerland': [],
    }
    for item in series:
        for key in result:
            if key in item:
                result[key].append(item)
    return result
                
```

```python
scrp = scraping(positions['text'])
```

```python
scrap = pd.DataFrame(scrp['PhD'], columns=['PhD'])
```

```python
for item in scrp:
    scrap[item] = pd.Series(scrp[item])
```
